```python
tickers = pd.read_csv("/Users/simonevanno/Desktop/centrale marseille/cours/3A/TEMPS2/Gestion Portefeuille/Projet/datas/Euronext_Equities_2022-12-02.csv", sep=";")
tickers_amsterdam = tickers[tickers['Currency']=='EUR']['Symbol'].tolist()
for i in range(len(tickers_amsterdam)):
    tickers_amsterdam[i] = tickers_amsterdam[i] + ".AS"
print("Nous avons :",len(tickers_amsterdam), " actions")
```

Récupération des données

```python
tickers = yf.Tickers(tickers_amsterdam)
datas = tickers.history(period='max')
datas.index = pd.to_datetime(datas.index)
```

```python
tickers = yf.Tickers(tickers_amsterdam)
datas_price = datas['Close'].pct_change()
datas_price = datas_price.loc[:, ~datas_price.iloc[-1].isna()] # Supprimer les colonnes ou y'a un NaN en dernière valeur
datas_price
```

```python
datas_volume = datas['Volume']
date_string_begin = '2018-01-01'
date_string_end = '2020-01-01'
datas_volume = datas_volume.loc[datas_volume.index<=datetime.strptime(date_string_end, '%Y-%m-%d')] # On ne garde que les données jusqu'à 2020
datas_volume = datas_volume.loc[datas_volume.index>=datetime.strptime(date_string_begin, '%Y-%m-%d')] # On ne garde que les données jusqu'à 2020
datas_volume
```

Supression des NaN

```python
datas_volume = datas_volume.bfill()
datas_volume = datas_volume.ffill()
datas_volume = datas_volume.dropna(axis=1)
datas_volume
```

```python
datas_volume = datas_volume.resample('Y').sum()
print(datas_volume)
```
